```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("company_sales_data.csv")
```

```python
df
```

```python
# Line Plot total profit column
df['total_profit'].plot(kind='line') 

# Show the plot
plt.title('Total profit per month')
plt.xlabel('Index')
plt.ylabel('total profit')
plt.grid(True)
plt.show()
```

```python
# Create subplots
fig, axes = plt.subplots(1,1,figsize=(4, 4))  # 1 row, 3 columns

# Line plot
df['bathingsoap'].plot(ax=axes, title='Bathing soap')
df["facewash"].plot(ax=axes, title="Facewash" )
plt.tight_layout()
plt.show()
```
